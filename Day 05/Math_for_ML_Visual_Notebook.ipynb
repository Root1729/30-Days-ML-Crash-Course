## üìê 1. Linear Algebra: Vectors & Dot Product

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
a = np.array([2, 3])
b = np.array([1, 1])

# Plot vectors
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector B')
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.grid(True)
plt.legend()
plt.title('Vector Visualization')
plt.show()

# Dot product
dot_product = np.dot(a, b)
print("Dot Product of A and B:", dot_product)
```

## üîÅ 2. Calculus: Visualizing a Derivative

```python
import sympy as sp
sp.init_printing()

x = sp.symbols('x')
f = x**3
df = sp.diff(f, x)

display(sp.Eq(sp.Derivative(f, x), df))
```

## ‚öôÔ∏è 3. Gradient Descent on a Quadratic Function

```python
# Gradient descent demo
def f(x): return x**2
def df(x): return 2*x

x = 10
learning_rate = 0.2
steps = [x]

for _ in range(10):
    grad = df(x)
    x = x - learning_rate * grad
    steps.append(x)

plt.plot(np.linspace(-10, 10, 100), [f(i) for i in np.linspace(-10, 10, 100)], label='f(x)=x^2')
plt.scatter(steps, [f(i) for i in steps], color='red')
plt.plot(steps, [f(i) for i in steps], color='red', linestyle='--')
plt.title('Gradient Descent on x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## üé≤ 4. Probability Distributions

```python
# Normal distribution
import seaborn as sns

data = np.random.normal(loc=0, scale=1, size=1000)
sns.histplot(data, kde=True)
plt.title('Normal Distribution (Mean=0, SD=1)')
plt.show()
```
